```python
kitchen_mapping = {
    'USA_HYPER_EQUIPPED': 'HYPER_EQUIPPED',
    'USA_INSTALLED': 'INSTALLED',
    'USA_UNINSTALLED': 'NOT_INSTALLED',
    'USA_SEMI_EQUIPPED': 'SEMI_EQUIPPED',
}

df['kitchen'] =df.kitchen.replace(kitchen_mapping)
df.kitchen.value_counts(dropna=False)
```

```python
state_scaler= [None,"TO_BE_DONE_UP","TO_RESTORE","TO_RENOVATE","GOOD","JUST_RENOVATED","AS_NEW"]
peb_scaler = [None, 'G_F', 'G_C', 'G', 'F_E', 'F_D', 'F_C', 'F', 'E_D', 'E_C', 'E', 'D', 'C', 'B_A', 'B', 'A', 'A_A+', 'A+', 'A++']
kitchen_scaler = [None, 'NOT_INSTALLED', 'SEMI_EQUIPPED', 'HYPER_EQUIPPED', 'INSTALLED']


```

```python
target = ['price']
int_cols = ['typeofproperty', 'typeofsale', 'bedroomcount', 'bathroomcount', 'constructionyear']
bool_cols = ['garden', 'terrace', 'swimmingpool']
float_cols = ['livingarea', 'gardenarea']
ordinal = ['stateofbuilding', 'peb', 'kitchen' ]
categorical =['subtypeofproperty', 'province']

cols_to_keep = target + int_cols +bool_cols + float_cols + ordinal + categorical

len(cols_to_keep)
```

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('int_col', SimpleImputer(strategy="most_frequent"), int_cols),
        ('float_col', SimpleImputer(strategy="mean"), float_cols),
        ('bool_col', SimpleImputer(strategy='constant', fill_value=0.0), bool_cols),
        ("ordinal_col", OrdinalEncoder(categories=[state_scaler, peb_scaler, kitchen_scaler], handle_unknown="use_encoded_value", unknown_value=-1), ordinal),
        ('categorical_col', OneHotEncoder(sparse_output= False), categorical)
    ],
    remainder='passthrough'
)

X = preprocessor.fit_transform(df[cols_to_keep])
df_preprocessed = pd.DataFrame(X, columns = preprocessor.get_feature_names_out())
df_preprocessed.head()
```
