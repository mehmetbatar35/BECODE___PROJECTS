```python
import pandas as pd
import pickle

absolute_path = r"C:\Users\mehme\becode---\BECODE___PROJECTS\01.IMMOELIZA\02.Data_Analysis\df.pkl"

with open(absolute_path, 'rb') as f:
    df = pickle.load(f)


pd.set_option('display.max_columns', None)
```

```python
df.describe().T
```

```python
display(df.shape)
Q1 = df['bathroomcount'].quantile(0.1)
Q3 = df['bathroomcount'].quantile(0.90)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound, upper_bound)
df = df[(df['bathroomcount'] > lower_bound) & (df['bathroomcount'] < upper_bound)]
display(df.shape)
```
